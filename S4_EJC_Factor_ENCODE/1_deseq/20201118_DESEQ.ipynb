```r
suppressMessages(library(DESeq2))
```

```r
#4A3 knockdown vs scramble HEPG2
count_file <- "../0_download_run_salmon/20201117_EJC_factors_stop_collapsed_counts.tsv"
row_names <- "tx_id"
countdata <- read.csv(count_file, row.names=row_names, sep='\t')[ ,c('HepG2_ctrl_1', 'HepG2_ctrl_2', 'HepG2_EIF4A3_1', 'HepG2_EIF4A3_2')]
countdata <- round(countdata)
condition <- factor(c(rep("ctl", 2), rep("exp", 2)))
# Create a coldata frame and instantiate the DESeqDataSet. See ?DESeqDataSetFromMatrix
coldata <- data.frame(row.names=colnames(countdata), condition)
dds <- DESeqDataSetFromMatrix(countData=countdata, colData=coldata, design=~condition)
# Run the DESeq pipeline
dds <- DESeq(dds)
#filter for at least 30 counts between all datasets
dds <- dds[ rowSums(counts(dds)) > 30, ]
# Get differential expression results
res <- results(dds)
## Order by adjusted p-value
res <- res[order(res$padj), ]
## Merge with normalized count data
#res <- merge(as.data.frame(res), as.data.frame(counts(dds, normalized=TRUE)), by="row.names", sort=FALSE)
#names(res)[1] <- "Transcript"
## Write results
write.table(res, file='DESeq_results/HEPG2_EIF4A3.txt', quote=F, sep='\t', col.names = NA, row.names = TRUE)

#4A3 knockdown vs scramble K562
countdata <- read.csv(count_file, row.names=row_names, sep='\t')[ ,c('K562_ctrl_for_EIF4A3_1', 'K562_ctrl_for_EIF4A3_2', 'K562_EIF4A3_1', 'K562_EIF4A3_2')]
countdata <- round(countdata)
condition <- factor(c(rep("ctl", 2), rep("exp", 2)))
# Create a coldata frame and instantiate the DESeqDataSet. See ?DESeqDataSetFromMatrix
coldata <- data.frame(row.names=colnames(countdata), condition)
dds <- DESeqDataSetFromMatrix(countData=countdata, colData=coldata, design=~condition)
# Run the DESeq pipeline
dds <- DESeq(dds)
#filter for at least 30 counts between all datasets
dds <- dds[ rowSums(counts(dds)) > 30, ]
# Get differential expression results
res <- results(dds)
## Order by adjusted p-value
res <- res[order(res$padj), ]
## Merge with normalized count data
#res <- merge(as.data.frame(res), as.data.frame(counts(dds, normalized=TRUE)), by="row.names", sort=FALSE)
#names(res)[1] <- "Transcript"
## Write results
write.table(res, file='DESeq_results/K562_EIF4A3.txt', quote=F, sep='\t', col.names = NA, row.names = TRUE)

#MAGOH knockdown vs scramble HEPG2
countdata <- read.csv(count_file, row.names=row_names, sep='\t')[ ,c('HepG2_ctrl_1', 'HepG2_ctrl_2', 'HepG2_MAGOH_1', 'HepG2_MAGOH_2')]
countdata <- round(countdata)
condition <- factor(c(rep("ctl", 2), rep("exp", 2)))
# Create a coldata frame and instantiate the DESeqDataSet. See ?DESeqDataSetFromMatrix
coldata <- data.frame(row.names=colnames(countdata), condition)
dds <- DESeqDataSetFromMatrix(countData=countdata, colData=coldata, design=~condition)
# Run the DESeq pipeline
dds <- DESeq(dds)
#filter for at least 30 counts between all datasets
dds <- dds[ rowSums(counts(dds)) > 30, ]
# Get differential expression results
res <- results(dds)
## Order by adjusted p-value
res <- res[order(res$padj), ]
## Merge with normalized count data
#res <- merge(as.data.frame(res), as.data.frame(counts(dds, normalized=TRUE)), by="row.names", sort=FALSE)
#names(res)[1] <- "Transcript"
## Write results
write.table(res, file='DESeq_results/HEPG2_MAGOH.txt', quote=F, sep='\t', col.names = NA, row.names = TRUE)

#MAGOH knockdown vs scramble K562
countdata <- read.csv(count_file, row.names=row_names, sep='\t')[ ,c('K562_ctrl_for_MAGOH_1', 'K562_ctrl_for_MAGOH_2', 'K562_MAGOH_1', 'K562_MAGOH_2')]
countdata <- round(countdata)
condition <- factor(c(rep("ctl", 2), rep("exp", 2)))
# Create a coldata frame and instantiate the DESeqDataSet. See ?DESeqDataSetFromMatrix
coldata <- data.frame(row.names=colnames(countdata), condition)
dds <- DESeqDataSetFromMatrix(countData=countdata, colData=coldata, design=~condition)
# Run the DESeq pipeline
dds <- DESeq(dds)
#filter for at least 30 counts between all datasets
dds <- dds[ rowSums(counts(dds)) > 30, ]
# Get differential expression results
res <- results(dds)
## Order by adjusted p-value
res <- res[order(res$padj), ]
## Merge with normalized count data
#res <- merge(as.data.frame(res), as.data.frame(counts(dds, normalized=TRUE)), by="row.names", sort=FALSE)
#names(res)[1] <- "Transcript"
## Write results
write.table(res, file='DESeq_results/K562_MAGOH.txt', quote=F, sep='\t', col.names = NA, row.names = TRUE)
```
