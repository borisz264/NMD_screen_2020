```python
def get_nmd_status(x):
    if '|' in x:
        return get_nmd_status(x.split('|')[0])
    return sorted(comprehensive_gtf_data.transcript_to_entries[x])[0].get_value('transcript_type') == 'nonsense_mediated_decay'

def get_gene_id(x):
    if '|' in x:
        return get_gene_id(x.split('|')[0])
    return comprehensive_gtf_data.tx_to_genes[x]

def get_gene_name(x):
    if '|' in x:
        return get_gene_name(x.split('|')[0])
    return comprehensive_gtf_data.tx_to_gene_names[x]

def log_trans(x):
    try:
        return math.log(x, 10)
    except:
        return float('NaN')
```

### now combining transcripts by stop codon

```python
def get_genomic_stop_coords(x):
    #if given a |-seperated list of tx, will only return value for first one
    if '|' in x:
        return get_genomic_stop_coords(x.split('|')[0])
    stop_entry = comprehensive_gtf_data.sorted_exons(x, exon_type='stop_codon')
    if len(stop_entry) == 0:
        return 'not annotated'
    stop = '%s %s %s %s' % (stop_entry[0].get_value('chr'), stop_entry[0].get_value('strand'), 
                             stop_entry[0].get_value('start'), stop_entry[0].get_value('end'))
    return stop
```

```python
combined_read_counts_df['stop_codon'] = combined_read_counts_df['tx_id'].apply(get_genomic_stop_coords)
combined_read_TPM_df['stop_codon'] = combined_read_TPM_df['tx_id'].apply(get_genomic_stop_coords)
```

```python
def format_for_prism(subset_df, output_prefix):
    treatments = ['DMSO', 'SMG1i']
    genotypes = ['WT', 'eIF4E2_4', 'GIGYF2_4']
    replicates = ['1', '2']
    molecules = ['RNA', 'mono', 'di']
    headers = []
    for treatment in treatments:
        for genotype in genotypes:
            for replicate in replicates:
                headers.append('%s_%s_%s' % (genotype, treatment, replicate))
    for genotype in ['NT1', 'SMG1i', 'eIF4E2', 'GIGYF2']:
        for replicate in replicates:
            headers.append('%s_%s' % (genotype, replicate))    
    print headers
    
    for tx_id in subset_df['tx_id'].unique():
        print tx_id
        single_tx =  subset_df[subset_df['tx_id'] == tx_id]
        out_file=open('%s_%s.tsv' % (output_prefix, tx_id.replace('|','_')), 'w')
        out_file.write('%s_%s\t%s\n' % (tx_id, single_tx['stop_codon'].tolist()[0],'\t'.join(headers)))
        for molecule in molecules:
            data = []
            for header in headers:
                if '%s_%s' % (header, molecule) in single_tx.columns:
                    data.append(single_tx['%s_%s' % (header, molecule)].tolist()[0])
            out_file.write('%s\t%s\n' % (molecule, '\t'.join(['%f' % datum for datum in data])))        
        out_file.close()

format_for_prism(combined_read_TPM_df[combined_read_TPM_df['gene_name']=='UGAC_reporter'], 'tpms_for_prism/20200823_UGAC_reporter_TPMs')
format_for_prism(combined_read_TPM_df[combined_read_TPM_df['gene_name']=='CCNL1'], 'tpms_for_prism/20200823_CCNL1_TPMs')
format_for_prism(combined_read_TPM_df[combined_read_TPM_df['gene_name']=='TAF1D'], 'tpms_for_prism/20200823_TAF1D_TPMs')
format_for_prism(combined_read_TPM_df[combined_read_TPM_df['gene_name']=='GIGYF2'], 'tpms_for_prism/20200823_GIGYF2_TPMs')
format_for_prism(combined_read_TPM_df[combined_read_TPM_df['gene_name']=='EIF4E2'], 'tpms_for_prism/20200823_EIF4E2_TPMs')
format_for_prism(combined_read_TPM_df[combined_read_TPM_df['gene_name']=='SRSF3'], 'tpms_for_prism/20200823_SRSF3_TPMs')
```

```python
def format_for_prism_normed(subset_df, output_prefix):
    treatments = ['DMSO', 'SMG1i']
    genotypes = ['WT', 'eIF4E2_4', 'GIGYF2_4']
    replicates = ['1', '2']
    molecules = ['RNA', 'mono', 'di']
    headers = []
    for treatment in treatments:
        for genotype in genotypes:
            for replicate in replicates:
                headers.append('%s_%s_%s' % (genotype, treatment, replicate)) 
    
    for molecule in molecules:
        out_file=open('%s_%s.tsv' % (output_prefix, molecule), 'w')
        out_file.write('%s\t%s\n' % ('tx_id', '\t'.join(headers)))
        for tx_id in subset_df['tx_id'].unique():
            single_tx =  subset_df[subset_df['tx_id'] == tx_id]
            NT_average = (single_tx['WT_DMSO_1_%s' % (molecule)].tolist()[0]+single_tx['WT_DMSO_2_%s' % (molecule)].tolist()[0])/2.0
            data = []
            for header in headers:
                if '%s_%s' % (header, molecule) in single_tx.columns:
                    if NT_average == 0:
                        data.append(float('inf'))
                    else:
                        data.append(single_tx['%s_%s' % (header, molecule)].tolist()[0]/NT_average)
            out_file.write('%s\t%s\n' % (tx_id, '\t'.join(['%f' % datum for datum in data])))        
        out_file.close()
```

```python
def format_for_prism_counts(subset_df, output_prefix):
    treatments = ['DMSO', 'SMG1i']
    genotypes = ['WT', 'eIF4E2_4', 'GIGYF2_4']
    replicates = ['1', '2']
    molecules = ['RNA', 'mono', 'di']
    headers = []
    for treatment in treatments:
        for genotype in genotypes:
            for replicate in replicates:
                headers.append('%s_%s_%s' % (genotype, treatment, replicate)) 
    
    for molecule in molecules:
        out_file=open('%s_%s.tsv' % (output_prefix, molecule), 'w')
        out_file.write('%s\t%s\n' % ('tx_id', '\t'.join(headers)))
        for tx_id in subset_df['tx_id'].unique():
            single_tx =  subset_df[subset_df['tx_id'] == tx_id]
            data = []
            for header in headers:
                if '%s_%s' % (header, molecule) in single_tx.columns:
                    data.append(single_tx['%s_%s' % (header, molecule)].tolist()[0])
            out_file.write('%s\t%s\n' % (tx_id, '\t'.join(['%f' % datum for datum in data])))        
        out_file.close()
```

```python
#fraction of reads for each gene mapping to transcripts without an annotated stop codon
from collections import defaultdict
datasets = list(combined_read_counts_df.columns[1:-3])
print datasets
fractions_mapping_to_unannotated = defaultdict(list)
number_mapping_to_unannotated = defaultdict(float)
total_mapping = defaultdict(float)


gene_grouped_counts = combined_read_counts_df.groupby('gene_id')
tx_grouped_counts = combined_read_counts_df.groupby('tx_id')

for gene_id in combined_read_counts_df['gene_id'].unique():
    gene_transcript_data = grouped_counts.get_group(gene_id)
    tx_ids = gene_transcript_data['tx_id'].unique()
    for dataset in datasets:
        total_gene_mapping = 0.
        unannotated_mapping = 0.
        for tx_id in tx_ids:
            tx_data = tx_grouped_counts.get_group(tx_id)
            #print(tx_data[dataset].tolist()[0])
            counts = float(tx_data[dataset].tolist()[0])
            total_gene_mapping += counts
            if get_genomic_stop_coords(tx_id)=='not annotated':
                unannotated_mapping += counts
        if total_gene_mapping > 0:
            total_mapping[dataset] += total_gene_mapping
            number_mapping_to_unannotated[dataset] += unannotated_mapping
            fractions_mapping_to_unannotated[dataset].append(float(unannotated_mapping)/float(total_gene_mapping))

for dataset in datasets:
    print(dataset, 'fraction_mapping_no_stop:', number_mapping_to_unannotated[dataset]/total_mapping[dataset])
```
