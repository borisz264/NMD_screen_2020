First I need to assemble the list of FASTQ files

```python
import os
fastq_dir = 'fastq/'
fastq_files = sorted([file_name for file_name in os.listdir(fastq_dir) if file_name.endswith('.fastq.gz') and not file_name.startswith('.')])
sample_names = ['_'.join(file_name[:-9].split('_')[2:]) for file_name in fastq_files]
fastq_full_path = [os.path.join(fastq_dir, file_name) for file_name in fastq_files]
```
