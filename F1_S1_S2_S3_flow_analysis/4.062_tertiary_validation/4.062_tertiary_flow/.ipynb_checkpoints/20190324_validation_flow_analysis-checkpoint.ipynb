```python
import fcsparser

import pandas as pd
import math
import os
#import all data, and merge into a single datatframe, with sample names in extra columns


def log_trans(x):
    try:
        return math.log(x, 10)
    except:
        return float('NaN')
    
cell_lines = ['cBZ056','cBZ058']
stop_contexts = ['UGAC','AQP4']
#guides = ['NT_1','NT_2','ZNF598_1','ZNF598_2','ASCC3_1','ASCC3_2','GIGYF2_1','GIGYF2_2']
data_folder = 'FCS/Boris/20190321_screen_validation'

#collect guide names from plate layout data
layout_file = open('20190313_transfection_layout.txt')
wells = []
guides = []
genes = []
guide_numbers = []
for line in layout_file.readlines()[1:]:
    well, guide_description = line.strip().split('\t')[:2]
    gene = guide_description.split('_')[1]
    guide_num = guide_description.split('_')[0]
    wells.append(well)
    guides.append('%s_%s' % (gene, guide_num))
    genes.append(gene)
    guide_numbers.append(guide_num)
layout_file.close()

#will only look at data gated on good scatter

dfs = []
index2 = 0
cell_line_index=0
for cell_line in cell_lines:
    index1 = 0
    for guide in guides:
        index1 += 1
        index2 += 1
        data_file_name = '%s_tdTomato_tagBFP_%03d_%03d.fcs' % (cell_line, index1, index2)
        data_file_path = os.path.join(data_folder, data_file_name)
        meta, data = fcsparser.parse(data_file_path, meta_data_only=False, reformat_meta=True)
        data['cell line'] = cell_line
        data['stop context'] = stop_contexts[cell_line_index]
        data['gene'] = genes[index1-1]
        data['guide'] = guides[index1-1]
        data['guide number'] = guide_numbers[index1-1]
        data['well'] = wells[index1-1]
        
        data['sample'] = '%s %s' % (stop_contexts[cell_line_index], guide)
        #dfs.append(data.sample(n=1000)) 
        dfs.append(data) #to subsample and speed up plotting, use the above line instead of this one
    cell_line_index+=1
all_data = pd.concat(dfs)
```

```python
#first just look at the forward and side scatter
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import gaussian_kde
import numpy as np
sns.set(style="white", color_codes=True)
def density_colored_scatter(x, y, color, **kwargs):
    # Calculate the point density
    xy = np.vstack([x,y])
    z = gaussian_kde(xy)(xy)
    plt.scatter(x, y, c=z, **kwargs)
    #plt.yscale('log')
    #plt.xscale('log')
    cutoff = 120000
    pos = [yval for yval in y if yval>=cutoff]
    percent_pos = 100.*len(pos)/len(y)
    plt.hlines(cutoff,0,25000, linestyle='dashed')
    plt.annotate("%.3f" % percent_pos, xy=(20000, cutoff+1))


g = sns.FacetGrid(all_data, col="sample", col_wrap=4)

cmap = 'viridis'
g= g.map(density_colored_scatter, "FSC-H", "FSC-W", cmap=cmap, s=1, linewidth=0)
plt.savefig('20190324_FSC_WH.pdf', transparent=True)
plt.savefig('20190324_FSC_WH.png', transparent=True)
```

```python
good_scatter = fsc_ssc_filter.rename(index=str, columns={'PE-A':'tdTomato-A', 'Pacific Blue-A':'tagBFP-A'})
good_scatter['tdTomato/tagBFP'] = good_scatter['tdTomato-A']/good_scatter['tagBFP-A']
good_scatter['log tdTomato'] = good_scatter['tdTomato-A'].apply(log_trans)
good_scatter['log tagBFP'] = good_scatter['tagBFP-A'].apply(log_trans)
good_scatter['log tdTomato/tagBFP'] = good_scatter['tdTomato/tagBFP'].apply(log_trans)
```

```python
grouped_data = good_scatter.groupby(['stop context', 'guide'])
```

```python
#colors - colorblind safe
black = (0,0,0)
gray = (0.6,0.6,0.6)
orange = (230/255.0,159/255.0,0)
skyBlue = (86/255.0,180/255.0,233/255.0)
bluishGreen = (0,158/255.0,115/255.0)
yellow = (240/255.0,228/255.0,66/255.0)
blue = (0,114/255.0,178/255.0)
vermillion = (213/255.0,94/255.0,0)
reddishPurple = (204/255.0,121/255.0,167/255.0)
colors = [black, vermillion, orange, skyBlue, bluishGreen, blue, reddishPurple, yellow, gray]
```

```python
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
import scipy.stats as stats

xmin=1
xmax=5
#ymin=-0.1
#ymax=4
plots=[]
num_plots_wide = 4
num_plots_high = 24

variable = 'log tagBFP'

fig = plt.figure(figsize=(2*num_plots_wide, 2*num_plots_high))
#fig.subplots_adjust(wspace=0.4, hspace=0.4)
plot_index =1
for guide in guides:
    plot = fig.add_subplot(num_plots_high, num_plots_wide, plot_index)
    stop_index = 0
    for stop_context in stop_contexts:
        nt1_data = grouped_data.get_group((stop_context, 'NT1'))
        nt2_data = grouped_data.get_group((stop_context, 'NT2'))
        guide_data = grouped_data.get_group((stop_context, guide))
        nt1_data[variable].plot.kde(color=colors[stop_index], linestyle=':', alpha=0.6, ax=plot, lw=2)
        nt2_data[variable].plot.kde(color=colors[stop_index], linestyle='--', alpha=0.6, ax=plot, lw=2)
        guide_data[variable].plot.kde(color=colors[stop_index], linestyle='-', alpha=1.0, ax=plot, lw=2)
        stop_index += 1
    plot_index += 1
    plot.set_title(guide)
    plot.set_xlabel(variable)
    plot.set_xlim(xmin, xmax)
    if plot_index%num_plots_wide==0:
        plot.set_ylabel('density')
plt.tight_layout()    
plt.savefig('20190324_tagBFP_kde.pdf', transparent=True)
```

```python
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
import scipy.stats as stats

xmin=-2
xmax=2
#ymin=-0.1
#ymax=4
plots=[]
num_plots_wide = 4
num_plots_high = 24

variable = 'log tdTomato/tagBFP'

fig = plt.figure(figsize=(2*num_plots_wide, 2*num_plots_high))
#fig.subplots_adjust(wspace=0.4, hspace=0.4)
plot_index =1
for guide in guides:
    plot = fig.add_subplot(num_plots_high, num_plots_wide, plot_index)
    stop_index = 0
    for stop_context in stop_contexts:
        nt1_data = grouped_data.get_group((stop_context, 'NT1'))
        nt2_data = grouped_data.get_group((stop_context, 'NT2'))
        guide_data = grouped_data.get_group((stop_context, guide))
        nt1_data[variable].plot.kde(color=colors[stop_index], linestyle=':', alpha=0.6, ax=plot, lw=2)
        nt2_data[variable].plot.kde(color=colors[stop_index], linestyle='--', alpha=0.6, ax=plot, lw=2)
        guide_data[variable].replace(-np.inf, np.nan).replace(np.inf, np.nan).dropna().plot.kde(color=colors[stop_index], linestyle='-', alpha=1.0, ax=plot, lw=2)
        stop_index += 1
    plot_index += 1
    plot.set_title(guide)
    plot.set_xlabel(variable)
    plot.set_xlim(xmin, xmax)
    plot.set_ylabel('')
plt.tight_layout()    
plt.savefig('20190324_ratio_kde.pdf', transparent=True)
```

```python
#compute median and standard deviation for each
import pandas as pd
from collections import defaultdict

data=defaultdict(list)
for guide in guides:
    for stop_context in stop_contexts:
        guide_data = grouped_data.get_group((stop_context, guide))
        data['tagBFP median'].append(guide_data['log tagBFP'].median())
        data['tagBFP standard deviation'].append(guide_data['log tagBFP'].std())
        data['tdTomato median'].append(guide_data['log tdTomato'].median())
        data['tdTomato standard deviation'].append(guide_data['log tdTomato'].std())
        data['guide'].append(guide)
        data['stop context'].append(stop_context)
summary_dataframe=pd.DataFrame(data=data)
summary_dataframe.to_csv('20190324_median_summary.txt', sep='\t')
```
